```
import keras
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('/content/drive/My Drive/Housing/train_final.csv')
df.head()
```

```
squareFeet = df[['INT_SQFT']].values #x value
salePrice = df[['SALES_PRICE']].values #y value
```

```
model = keras.Sequential()
model.add(keras.layers.Dense(1, input_shape=(1,)))
model.compile(keras.optimizers.Adam(lr=1), 'mean_squared_error')

model.fit(squareFeet,salePrice, epochs=30, batch_size=10)
```

```
plt.scatter(squareFeet,salePrice, label='Housing', color='red')

plt.xlabel('Squarefeet')
plt.ylabel('Saleprice')
plt.title('House Pricing')


y_pred = model.predict(squareFeet) #The predicted housing price based on square feet

#Plot the linear regression line
plt.plot(squareFeet, y_pred, color='blue')
```
